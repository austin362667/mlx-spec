# Train a draft model

1. Initialize the draft model based on [`Qwen/Qwen3-0.6B-MLX-bf16`](https://huggingface.co/Qwen/Qwen3-0.6B-MLX-bf16)
2. Prepare the training dataset (Python coding tasks, e.g.,[`iamtarun/python_code_instructions_18k_alpaca`](https://huggingface.co/datasets/iamtarun/python_code_instructions_18k_alpaca)).
    - Resample the dataset by regenerating responses using `Qwen3-4B`.
    - Ref: Regenerate Datasets in [SpecForge](https://docs.sglang.io/SpecForge/basic_usage/data_preparation.html#regenerate-datasets)
    - Save regenerated Parquet file at [`Austin362667/python_code_instructions_5k_alpaca_qwen3_4B_resampled`](https://huggingface.co/datasets/Austin362667/python_code_instructions_5k_alpaca_qwen3_4B_resampled)

3. Perform full-weight fine-tuning with KL-divergence loss at `./train.py`:
    - Similar to a knowledge distillation setup.
    - Minimize the distance between the output distributions of the teacher model and the student model (the draft model in this case).
